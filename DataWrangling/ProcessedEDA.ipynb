```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('Processed.csv')

df.head()
```

```python
# Histogram of Vote Count
fig = px.histogram(df,
                   x="vote_average",
                   facet_col="Is Christmas Movie",
                   title='Vote Average',
                   color_discrete_sequence=px.colors.qualitative.Dark24)
fig.show()
```

```python
# Box Plot of Vote Count
fig = px.box(df,
                   x="vote_average",
                   color="Is Christmas Movie",
                   title='Vote Average',
                   color_discrete_sequence=px.colors.qualitative.Dark24)
fig.show()
```
